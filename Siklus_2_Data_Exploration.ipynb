# Data Exploration

## Summary Descriptive Statistics

Summary descriptive statistics adalah rangkuman properti dari sebuah data

Properti tersebut adalah berupa:
- frekuensi,
- modus,
- mean,
- range, dan
- standar deviasi (deviation standard).

Tampilkan informasi data menggunakan syntax berikut:

image.png

```python
df.info()
```

Berdasarkan informasi yang ditampilkan, data memiliki:
- 67 baris
- 5 kolom:
    - Sex dengan tipe data object/str
    - Weight (Sep) dengan tipe data int
    - Weight (Apr) dengan tipe data int
    - BMI (Sep) dengan tipe data float
    - BMI (Apr) dengan tipe data float

---

Pada Python, kita tidak perlu membuat program untuk frekuensi, mean, standar deviasi, range, dan kuartil.

Kita hanya perlu memanggil method `describe()`.


Method `describe()` secara otomatis menghitung:
- frekuensi (count)
- mean
- median (kuartil 50%)
- standar deviasi
- range (nilai min dan max)
- kuartil

Kolom-kolom yang **datanya numerik**.

image.png

```python
df.describe()
```

Coba simpulkan hasil `describe()` di atas!



---

Selain membuat deskripsi kolom numerik, kita juga bisa membuat deskripsi kolom kategorik dengan menambahkan argumen `include='all'` pada `describe()`.



image.png

```python
df.describe(include='object')
```

Coba simpulkan hasil `describe(include='object')` di atas!

## Data Visualization

Data Visualization (visualisasi data) adalah tampilan berupa grafis atau visual dari informasi dan data.

Dalam kata lain, data visualization mengubah kumpulan data menjadi hal lebih sederhana untuk ditampilkan.

Dengan menggunakan elemen visual tersebut, pembaca akan lebih mudah memahami tren, pola, dan outliers dalam suatu data.

Pada Python, terdapat dua library yang umum digunakan untuk membuat Data Visualization, yakni:
1. [Matplotlib](https://matplotlib.org/stable/index.html)
2. [Seaborn](https://seaborn.pydata.org/)

### Visualisasi Data Menggunakan Library Matplotlib

image.png

```python
import matplotlib.pyplot as plt
```

#### 1. Bar Graph atau Bar Chart

Biasa digunakan untuk membandingkan kategori.

image.png

```python
# Menghitung jumlah pria dan wanita
jumlah_pria = df['Sex'].value_counts()['M']
jumlah_wanita = df['Sex'].value_counts()['F']

# Membuat Bar Graph
plt.bar(['Pria', 'Wanita'], [jumlah_pria, jumlah_wanita])
plt.show()
```

Kita dapat mengubah ukuran dan menambah keterangan lain dalam plot, seperti:
- Judul plot
- Nama sumbu x
- Nama sumbu y
- Warna Bar Chart

image.png

```python
plt.figure(figsize=(10, 3))
plt.bar(['Pria', 'Wanita'], [jumlah_pria, jumlah_wanita], color=['orange', 'pink'])
plt.title('Perbandingan jumlah Pria dan Wanita')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.show()
```

Contoh bar graph/chart lainnya:



image.png

```python
# Menghitung rata-rata berat Pria dan Wanita bulan September
rata2berat_pria = df[df['Sex']=='M'][' "Weight (Sep)"'].mean()
rata2berat_wanita = df[df['Sex']=='F'][' "Weight (Sep)"'].mean()
weights = [rata2berat_pria, rata2berat_wanita]
labels = ['Pria', 'Wanita']

# Membuat Bar Graph
plt.figure(figsize=(7, 5))
plt.bar(labels, weights)
plt.title('Perbandingan rata-rata berat badan pria dan wanita bulan September')
plt.show()
```

#### 2. Histogram

Biasa digunakan untuk menampilkan distribusi frekuensi.

Jalankan program berikut kemudian simpulkan plot-nya!

image.png

```python
plt.hist(df[' "Weight (Sep)"'])
plt.title('Histogram dari berat badan bulan April')
plt.ylabel('Frekuensi')
plt.xlabel('Berat')
plt.show()
```

Cara lain dalam membuat histogram:

Jalankan program lalu simpulkan plot-nya!

image.png

```python
df[' "Weight (Apr)"'].plot(kind='hist', figsize=(7, 5))
plt.title('Histogram dari berat badan bulan April')
plt.ylabel('frekuensi')
plt.xlabel('berat')
plt.show()
```

Apa perbedaan antara bar graph dan histogram?

#### 3. Line Chart

Biasa digunakan untuk menunjukkan tren pada data atau membandingkan data.

Contoh line chart:

image.png

```python
bmi_april = df[' "BMI (Sep)"']

plt.plot(bmi_april)
plt.title('Perbandingan BMI bulan April dan September')
plt.xlabel('Peserta')
plt.ylabel('BMI')
plt.show()
```

Contoh line chart untuk membandingkan data:



image.png

```python
weight_september = df[' "Weight (Sep)"']
bmi_september = df[' "BMI (Sep)"']

plt.plot(weight_september)
plt.plot(bmi_september)
plt.title('Perbandingan Berat dan BMI bulan September', fontsize = 10)
plt.xlabel('Peserta', fontsize = 10)
plt.ylabel('Nilai', fontsize = 10)
plt.show()
```

Kita dapat menambahkan legenda dalam plot.



image.png

```python
weight_september = df[' "Weight (Sep)"']
bmi_september = df[' "BMI (Sep)"']

plt.plot(weight_september, label='Berat bulan September')
plt.plot(bmi_september, label='BMI bulan September')
plt.title('Perbandingan Berat dan BMI bulan September', fontsize=10)
plt.xlabel('Peserta', fontsize = 10)
plt.ylabel('Nilai', fontsize = 10)
plt.legend()
plt.show()
```

Kita juga dapat mengubah bentuk dan warna garis dalam plot:

image.png

```python
weight_september = df[' "Weight (Sep)"']
bmi_september = df[' "BMI (Sep)"']

plt.plot(weight_september, 'o--', label='Berat bulan September')
plt.plot(bmi_september, 'r--', label='BMI bulan September')
plt.title('Perbandingan Berat dan BMI bulan September', fontsize=10)
plt.xlabel('Peserta', fontsize = 10)
plt.ylabel('Nilai', fontsize = 10)
plt.legend()
plt.show()
```

Cara lain dalam membuat line chart:


image.png

```python

df[[' "BMI (Apr)"', ' "BMI (Sep)"']].plot(kind='line', figsize=(10, 5))
plt.title('Perbandingan BMI bulan April dan September', fontsize=10)
plt.xlabel('Peserta', fontsize = 10)
plt.ylabel('BMI', fontsize = 10)
plt.legend()
plt.show()
```

#### 4. Scatter plot

Menyajikan hubungan antara dua variabel dalam kumpulan data.

Contohnya menampilkan hubungan antara Berat dan BMI bulan April:

image.png

```python
weight_september = df[' "Weight (Apr)"']
bmi_september = df[' "BMI (Apr)"']


plt.scatter(weight_september, bmi_september)
plt.title('Perbandingan Berat dan BMI bulan April')
plt.xlabel('Berat', fontsize = 10)
plt.ylabel('BMI', fontsize = 10)
plt.legend()
plt.show()
```

Contoh lain dalam membuat scatter plot:

image.png

```python
df[[' "Weight (Sep)"', ' "Weight (Apr)"']]. plot(kind='scatter',
                                                 x=' "Weight (Sep)"',
                                                 y=' "Weight (Apr)"',
                                                 figsize=(10, 6),
                                                 color='red')
plt.title('Hubungan Berat dan BMI bulan April')
plt.xlabel('Berat', fontsize = 10)
plt.ylabel('BMI', fontsize = 10)
plt.show()
```

Scatter plotnya seperti membentuk garis lurus serong kanan. Artinya, hubungan berat badan bulan April dan September erat dan positif.

#### 5. Pie Chart

Menunjukkan proporsi relasional antara data.

image.png

```python
jumlah_pria = df['Sex']. value_counts()['M']
jumlah_wanita = df['Sex'].value_counts()['F']
counts = [jumlah_pria, jumlah_wanita]
labels = ['Pria', 'Wanita']
colors = ['lightblue', 'pink']

plt.figure(figsize =(7, 5))
plt.pie(counts, labels=labels, colors=colors, autopct='%.2f')
plt.title('Perbandingan jumlah Pria dan Wanita')
plt.show
```

#### 6. Heatmap (Korelasi)

heatmap.PNG

```python
# Convert 'Sex' column to numerical representation using one-hot encoding
df = pd.get_dummies(df, columns=['Sex'], drop_first=True) # drop_first to avoid multicollinearity

# Calculate correlation matrix
corrmat = df.corr()

# Plot heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(corrmat, annot=True)
```

Warna merepresentasikan keeratan hubungan:
- bisa diliat dari nilai hubungan/korelasinya.
- bisa dilihat dari penggaris di sebelah heatmap.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = 'cities.csv'
data = pd.read_csv(file_path)
```

```python
# Bersihkan nama kolom
data.columns = data.columns.str.strip().str.replace('"', '')
```

```python
# 1. Summary Descriptive Statistics
summary_stats = data.describe(include='all')
print("Summary Descriptive Statistics:")
print(summary_stats)
```

```python
# 2. Kesimpulan dari Summary Descriptive Statistics
print("\nKesimpulan Summary Descriptive Statistics:")
print("""
1. Latitude (LatD):
   - Rata-rata: 38.82°, Rentang: 26° hingga 50°
   - Semua data berada di belahan bumi utara (kolom NS hanya berisi 'N').

2. Longitude (LonD):
   - Rata-rata: 93.25°, Rentang: 71° hingga 123°
   - Semua data berada di belahan bumi barat (kolom EW hanya berisi 'W').

3. Kota dan Negara Bagian:
   - Dataset mencakup 120 kota unik di 47 negara bagian.
   - Kota paling umum adalah "Springfield" (4 entri).
   - Negara bagian dengan entri terbanyak adalah California (CA) dengan 12 kota.
""")
```

```python
# 3. Visualisasi Data
sns.set(style="whitegrid")
```

```python
# Plot Distribusi Latitude dan Longitude Degrees
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.histplot(data['LatD'], kde=True, ax=axes[0], color="skyblue")
axes[0].set_title("Distribution of Latitude Degrees")
axes[0].set_xlabel("Latitude Degree")

sns.histplot(data['LonD'], kde=True, ax=axes[1], color="salmon")
axes[1].set_title("Distribution of Longitude Degrees")
axes[1].set_xlabel("Longitude Degree")

plt.tight_layout()
plt.show()
```

```python
# Plot Top 10 Negara Bagian dengan Jumlah Kota Terbanyak
top_states = data['State'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_states.values, y=top_states.index, palette="viridis")
plt.title("Top 10 States by Number of Cities")
plt.xlabel("Number of Cities")
plt.ylabel("State")
plt.show()
```

```python
# 4. Kesimpulan Visualisasi
print("\nKesimpulan Visualisasi Data:")
print("""
- Distribusi Latitude menunjukkan banyak kota berada di sekitar 35-45° LU.
- Distribusi Longitude menunjukkan sebagian besar kota berada di sekitar 70-100° BB.
- Negara bagian dengan jumlah kota terbanyak adalah California, disusul oleh Texas dan Ohio.
""")
```
